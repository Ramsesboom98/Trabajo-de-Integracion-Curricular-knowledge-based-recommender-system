```python
import sqlite3
import xml.etree.ElementTree as ET
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
v30_ni = pd.read_csv('compounds30_ni_29.csv',index_col=0)
v30_ni.columns =['compuesto','cell','active']
v30_ni
```

```python
df_rec = pd.read_csv("recomendaciones_top10.csv").drop_duplicates(keep='first')
df_count = df_rec.groupby(by=['compuesto','cell'])
index1 = pd.MultiIndex.from_arrays([df_rec[col] for col in ['compuesto', 'cell']])
index2 = df_count.size()[df_count.size()==1].index
df_rec=df_rec.loc[index1.isin(index2)]
df_rec
```

```python
merged_df = pd.merge(df_rec, v30_ni,  how='left', left_on=['compuesto','cell'], right_on = ['compuesto','cell'])
merged_df.columns=['compuesto','cell','prediccion','realidad']
merged_df
```

```python
merged_df.to_csv('comparaciones_v1.csv',header=True,index=True)
```

```python
df_comparaciones = pd.read_csv("comparaciones_v1.csv",index_col=0)
df_comparaciones
```

```python
# COMPARACION TODOS LOS VALORES EN UN CSV * COMPUESTO CELL PREDICCION REALIDAD 
# 1 VERDADERO POSITIVO  prediccion 1 realidad 1 * contar y dividir para el numero total de los que enrealidad son 1
# 2 falso positivo  predije -1  y la realidad es 1 * contar y dividir para el numero total de los que enrialidad son 1 
# 3 verdadero negativo predije -1 y la realidad es -1 * contar y dividir para el numero total de los que enrialidad son -1
# 4 falso negativo predije 1 y la realidad es -1 * contar y dividir para el numero total de los que enrialidad son -1 
# porcentajes 
# libreria sklearn * confusion matrix 
```

```python
from sklearn.metrics import confusion_matrix
y_true = df_comparaciones['prediccion'].tolist()
y_pred = df_comparaciones['realidad'].tolist()
confusion_matrix(y_pred, y_true)
```
